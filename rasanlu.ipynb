```python
from rasa_nlu.training_data import load_data
from rasa_nlu.model import Trainer
from rasa_nlu import config
from rasa_nlu.model import Interpreter

def train_horoscope_bot(data_jason, config_file, model_dir):
    #Run the data stored in the file passed as a parameter
    training_data = load_data(data_jason)

    #This two commands trains the data based on the pipeline 'tensorflow_embedding'"
    trainer = Trainer(config.load(config_file))
    trainer.train(training_data)

    
    model_directory = trainer.persist(model_dir, fixed_model_name= "horoscopebot")

def predict_intent(text):
    interpreter = Interpreter.load("./models/nlu/default/horoscopebot")
    print(interpreter.parse(text))

```

```python
train_horoscope_bot('./data/data.json', "config.json", "./models/nlu")
```

```python
predict_intent("I'm lookign for my horoscope for today. I am wondering if you can tell me that.")
```
